Полиномиальная регрессия

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def solve(n):
    points = np.linspace(1, 15, n)
    r=np.linalg.solve(np.array([[x**i for i in range(n)] for x in points]), f(points))
    return np.array([sum([r[i]*x**i for i in range(n)]) for x in points])

                                                   
xx = np.linspace(1, 15)
yy = f(xx)
plt.plot(xx, yy, color = 'r')
xx = np.linspace(1, 15,5)
yy = f(xx)
plt.plot(xx, yy,'.', color = 'b')
xx = np.linspace(1, 15,5)
yy = solve(5)
plt.plot(xx, yy, color = 'k')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('Первый случай n=5')
plt.grid()
plt.show()

xx = np.linspace(1, 15)
yy = f(xx)
plt.plot(xx, yy, color = 'r')
xx = np.linspace(1, 15,15)
yy = f(xx)
plt.plot(xx, yy,'.', color = 'b')
xx = np.linspace(1, 15,15)
yy = solve(15)
plt.plot(xx, yy, color = 'k')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('Второй случай n=15')
plt.grid()
plt.show()

xx = np.linspace(1, 15)
yy = f(xx)
plt.plot(xx, yy, color = 'r')
xx = np.linspace(1, 15,100)
yy = f(xx)
plt.plot(xx, yy,'.', color = 'b')
xx = np.linspace(1, 15,100)
yy = solve(100)
plt.plot(xx, yy, color = 'k')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('Третий случай n=100')
plt.grid()
plt.show()

print('Чем больше кол-во точек, тем лучше полином приближает исходную функцию. Полином ведет себя лучше всего в последнем случае')
```

Полиномиальная регрессия с зашумлённым данными 

```python

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def solve(n):
    points=np.linspace(1, 15, n)
    r=np.linalg.solve(np.array([[x**i for i in range(n)] for x in points]), f(points)+np.random.normal(0, 0.01, n))
    return np.array([sum([r[i]*x**i for i in range(n)]) for x in points])
                                                   
xx = np.linspace(1, 15)
yy = f(xx)
plt.plot(xx, yy, color = 'r')
xx = np.linspace(1, 15,5)
yy = f(xx)
plt.plot(xx, yy,'.', color = 'b')
xx = np.linspace(1, 15,5)
yy = solve(5)
plt.plot(xx, yy, color = 'k')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('Первый случай n=5')
plt.grid()
plt.show()

xx = np.linspace(1, 15)
yy = f(xx)
plt.plot(xx, yy, color = 'r')
xx = np.linspace(1, 15,15)
yy = f(xx)
plt.plot(xx, yy,'.', color = 'b')
xx = np.linspace(1, 15,15)
yy = solve(15)
plt.plot(xx, yy, color = 'k')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('Второй случай n=15')
plt.grid()
plt.show()

xx = np.linspace(1, 15)
yy = f(xx)
plt.plot(xx, yy, color = 'r')
xx = np.linspace(1, 15,100)
yy = f(xx)
plt.plot(xx, yy,'.', color = 'b')
xx = np.linspace(1, 15,100)
yy = solve(100)
plt.plot(xx, yy, color = 'k')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('Третий случай n=100')
plt.grid()
plt.show()

print('При увеличении числа точек шум очень сильно влияет на построение.Наблюдаются скачки.Полином ведет себя лучше всего во втором случае')

```

Метод наименьших квадратов 


```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def solve(n):
    points=np.linspace(1, 15, n)
    r,a,b,c=np.linalg.lstsq(np.array([[x**i for i in range(4)] for x in points]), f(points)+np.random.normal(0, 0.1,n),rcond=None)
    return np.array([sum([r[i]*x**i for i in range(4)]) for x in points]), f(points)+np.random.normal(0, 0.1,n)

plt.figure(figsize=(15, 10))                                                   
xx = np.linspace(1, 15,100)
yy = f(xx)
plt.plot(xx, yy, color = 'r')
xx = np.linspace(1, 15,100)
yy0,yy1 = solve(100)
plt.plot(xx, yy1,'.', color = 'b')
plt.plot(xx, yy0, color = 'k')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.grid()
plt.show()



print('Итог: полиномиальная регрессия без шумов дает хороший результат при большом кол-ве точек. С шумами наооборот: чем меньше, тем лучше. Метод наименьших квадратов дает хороший результат при кубическом полиноме')


```
