```python
import numpy as np
#from scipy.special import logsumexp
def f8(x):
    return (2*x-1)/(2*x+1)

def f9(x):
    #return logsumexp( 1/np.sqrt(2+3*x-2*x**2, where = x!=2))
    return 1/np.sqrt(2+3*x-2*x**2, where = x!=2)

def compute_integrate(f,a, b, eps, method):
    n = int((b-a) / eps)
    xx = np.linspace(a, b, n,endpoint=False)
    if method == 'rect':
        I=eps*np.sum(f(xx))
    elif method == 'trap':
        I=eps/2*(f(xx[0]) + 2*np.sum(f(xx[1:-1])) + f(xx[-1]))
    elif method == 'simpson':
        I=eps/3*(f(xx[0]) + 2*np.sum(f(xx[2:-1:2])) + 4*np.sum(f(xx[1:-2:2])) + f(xx[-1]))
    return I
        
    
print('интеграл 4:', compute_integrate(f8, 0, 2, 0.00001, 'rect') )
print('интеграл 4:', compute_integrate(f8, 0, 2, 0.00001, 'trap') )
print('интеграл 4:', compute_integrate(f8, 0, 2, 0.00001, 'simpson'))
print('интеграл 9:', compute_integrate(f9, 3/4, 2, 0.00000001, 'rect'))
print('интеграл 9:', compute_integrate(f9, 3/4, 2, 0.00000001, 'trap') )
print('интеграл 9:',compute_integrate(f9, 3/4, 2, 0.00000001, 'simpson'))
print('Я получила отличие в ответах в 5 знаке после запятой.\n Вывод: с уменьшением шага решения, ответ получается точнее ')




```
